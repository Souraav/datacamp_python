```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris_array = datasets.load_iris()
iris_measures = pd.DataFrame(iris_array['data'])
iris_species = pd.DataFrame(iris_array['target'])
iris = pd.concat([iris_measures.reset_index(drop=True), iris_species], axis=1)
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
iris.head()
```

```python
iris.plot(x='sepal_length', y='sepal_width', kind='scatter')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='box')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='hist')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='hist',
         bins=30, range=(4,8), normed=True)
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='hist', bins=30,
          range=(4,8), cumulative=True, normed=True)
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
# data prep
from urllib.request import urlretrieve
url = 'https://assets.datacamp.com/production/course_1639/datasets/auto-mpg.csv'
urlretrieve(url, 'auto-mpg.csv')
df_mpg = pd.read_csv('auto-mpg.csv')
df = df_mpg
print(df.shape)
print(df.head)
sizes = (df.weight/max(df.weight))*100
print(sizes)
```

```python
# Generate a scatter plot
df.plot(kind='scatter', x='hp', y='mpg', s=sizes)

# Add the title
plt.title('Fuel efficiency vs Horse-power')

# Add the x-axis label
plt.xlabel('Horse-power')

# Add the y-axis label
plt.ylabel('Fuel efficiency (mpg)')

# Display the plot
plt.show()
```

```python
# Make a list of the column names to be plotted: cols
cols = ['weight', 'mpg']

# Generate the box plots
df[cols].plot(kind='box', subplots=True)

# Display the ploat
plt.show()
```

```python
# Data prep - we can re-use df_mpg
df = df_mpg

# Compute the global mean and global standard deviation: global_mean, global_std
global_mean = df.mean()
global_std = df.std()

# Filter the US population from the origin column: us
us = df[df['origin'] == 'US']

# Compute the US mean and US standard deviation: us_mean, us_std
us_mean = us.mean()
us_std = us.std()

# Print the differences
print(us_mean - global_mean)
print(us_std - global_std)
```
