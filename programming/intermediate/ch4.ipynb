```python
# data prep
import pandas as pd
import numpy as np

baseball = pd.read_csv("baseball.csv")
height = baseball['Height'].tolist()
weight = baseball['Weight'].tolist()
h_w = list(zip(height, weight))
print(h_w[0:2])

np_height = np.array(height)
np_baseball = np.array(h_w)
```

```python
len(np_height)
```

```python
# Import numpy as np
import numpy as np

# For loop over np_height
# don't need np.nditer for vectors
for h in np_height:
    print(str(h) + " inches")

# For loop over np_baseball
for x in np.nditer(np_baseball):
    print(x)
```

```python
for lab, row in baseball.iterrows():
    print(lab)
    print(row)
```

```python
# lets create a new row in the loop, number of characters in the persons name
for lab, row in baseball.iterrows():
    baseball.loc[lab, "name_length"]  = len(row["Name"])

print(baseball.loc[:3,:])
```

```python
# apply version - much quicker and more data-science-thonic.
baseball = baseball.iloc[:, :7]
print(baseball.loc[:3,:])

baseball['name_length'] = baseball['Name'].apply(len)
print(baseball.loc[:3,:])
```
