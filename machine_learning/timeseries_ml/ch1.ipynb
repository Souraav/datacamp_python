```python
from sklearn.datasets import load_iris

iris = load_iris()
features = pd.DataFrame(iris.data, columns=iris.feature_names)
target = pd.Series(iris.target, name='target')
target = target.astype('category')
#target.cat.rename_categories(iris.target_names, inplace=True)
data = pd.concat([features, target], axis='columns')
data.info()
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

X = data.drop('target', axis='columns')
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=50)

# Construct data for the model
X = X_train[['petal length (cm)', 'petal width (cm)']]
y = y_train

# Fit the model
model = LinearSVC()
model.fit(X, y)

model.score(X,y)
```

```python
# Create input array
X_predict = X_test[['petal length (cm)', 'petal width (cm)']]

# Predict with the model
predictions = model.predict(X_predict)
print(predictions)

# Visualize predictions and actual values
plt.scatter(X_predict['petal length (cm)'], X_predict['petal width (cm)'],
            c=predictions, cmap=plt.cm.coolwarm)
plt.title("Predicted class values")
plt.show()
```

```python
# data prep
from sklearn.datasets import load_boston
boston_np = load_boston()
features = pd.DataFrame(boston_np.data, columns=boston_np.feature_names)
target = pd.Series(boston_np.target, name='target')
#target = target.astype('category')
#target.cat.rename_categories(iris.target_names, inplace=True)
boston = pd.concat([features, target], axis='columns')
boston.info()
```

```python
from sklearn import linear_model

# Prepare input and output DataFrames
X = boston[['AGE']]
y = boston[['RM']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=100)

# Fit the model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
model.score(y_test, X_test)
```

```python
# Generate predictions with the model using those inputs
predictions = model.predict(X_test)

# Visualize the inputs and predicted values
plt.scatter(y_test, predictions, color='r', s=3)
plt.xlabel('inputs')
plt.ylabel('predictions')
plt.show()
```
