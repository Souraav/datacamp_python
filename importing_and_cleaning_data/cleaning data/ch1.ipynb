```python
# data prep
from urllib.request import urlretrieve
url = 'https://assets.datacamp.com/production/course_2023/datasets/dob_job_application_filings_subset.csv'
urlretrieve(url, 'dob_job_application_filings_subset.csv')
```

```python
# Import pandas
import pandas as pd

# Read the file into a DataFrame: df
df = pd.read_csv('dob_job_application_filings_subset.csv')

# Print the head of df
print(df.head())

# Print the tail of df
print(df.tail())

# Print the shape of df
print(df.shape)

# Print the columns of df
print(df.columns)

# data prep
df_subset = df[['Job #', 'Doc #', 'Borough', 'Initial Cost', 'Total Est. Fee',
       'Existing Zoning Sqft', 'Proposed Zoning Sqft',
       'Enlargement SQ Footage', 'Street Frontage', 'ExistingNo. of Stories',
       'Proposed No. of Stories', 'Existing Height', 'Proposed Height']]

# Print the head and tail of df_subset
print(df_subset.head())
print(df_subset.tail())
print(df_subset.shape)
```

```python
# Print the info of df
print(df.info())

# Print the info of df_subset
print(df_subset.info())
```

```python
# Data prep
df['initial_cost'] = df['Initial Cost'].str.replace('$','')
df['initial_cost'].describe()
df['initial_cost'] = pd.to_numeric(df['initial_cost'])
```

```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt

# Create the boxplot
df.boxplot(column='initial_cost', by='Borough', rot=90)

# Display the plot
plt.show()
```

```python
# data prep
df['Total Est. Fee'].describe()
df['total_est_fee'] = df['Total Est. Fee'].str.replace('$','')
df['total_est_fee'].describe()
df['total_est_fee'] = pd.to_numeric(df['total_est_fee'])

df_subset['initial_cost'] = df_subset['Initial Cost'].str.replace('$','')
df_subset['initial_cost'].describe()
df_subset['initial_cost'] = pd.to_numeric(df['initial_cost'])

# My subset doesn't match his - he's removed his outliers, and uses all columns...
df_subset['Total Est. Fee'].describe()
df_subset['total_est_fee'] = df_subset['Total Est. Fee'].str.replace('$','')
df_subset['total_est_fee'].describe()
df_subset['total_est_fee'] = pd.to_numeric(df['total_est_fee'])
```
