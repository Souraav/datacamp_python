```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from urllib.request import urlretrieve
from io import StringIO
```

```python
file = 'young-people-survey-responses.csv'
url = ' https://assets.datacamp.com/production/repositories/3996/datasets/ab13162732ae9ca1a9a27e2efd3da923ed6a4e7b/'
urlretrieve(url+file, file)
survey_data = pd.read_csv(file)

survey_data['Age Category'] = '21+'
idx = survey_data['Age'] < 21
survey_data.loc[idx, 'Age Category'] = 'Less than 21'

idx = survey_data['Mathematics'] >=4
survey_data['Interested in Math'] = False
survey_data.loc[idx, 'Interested in Math'] = True
survey_data.info()
```

```python
survey_data['Age Category'].value_counts(dropna=False)
```

```python
# Create count plot of internet usage
sns.catplot(data=survey_data,
            x='Internet usage',
            kind='count')

# Show plot
plt.show()
```

```python
# Create count plot of internet usage
sns.catplot(data=survey_data,
            y='Internet usage',
            kind='count')

# Show plot
plt.show()
```

```python
# Create count plot of internet usage
sns.catplot(data=survey_data,
            y='Internet usage',
            col='Age Category',
            kind='count')

# Show plot
plt.show()
```

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(kind='bar',
            data=survey_data,
            x='Gender',
            y='Interested in Math')

# Show plot
plt.show()
```

```python
file = 'student-alcohol-consumption.csv'
url = 'https://assets.datacamp.com/production/repositories/3996/datasets/61e08004fef1a1b02b62620e3cd2533834239c90/'
urlretrieve(url+file, file)
student_data = pd.read_csv(file)
student_data.info()
```

```python
# Create bar plot of average final grade in each study category
sns.catplot(kind='bar',
            data=student_data,
            x='study_time',
            y='G3')

# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=['<2 hours', '2 to 5 hours', '5 to 10 hours', '>10 hours'])

# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=['<2 hours', '2 to 5 hours', '5 to 10 hours', '>10 hours'],
            ci=None)

# Show plot
plt.show()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(data=student_data,
            kind='box',
            x='study_time',
            y='G3',
            order=study_time_order)

# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(data=student_data,
            kind='box',
            x='internet',
            y='G3',
            hue='location',
            sym="")

# Show plot
plt.show()
```

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=0.5)

# Show plot
plt.show()
```

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point")
            
# Show plot
plt.show()
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2)
            
# Show plot
plt.show()
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            join=False,
            capsize=0.2)
            
# Show plot
plt.show()
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue='school')
            
# Show plot
plt.show()
```

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None)

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator=median)

# Show plot
plt.show()
```
