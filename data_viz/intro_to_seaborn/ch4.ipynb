```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from urllib.request import urlretrieve
from io import StringIO
```

```python
file = 'young-people-survey-responses.csv'
url = ' https://assets.datacamp.com/production/repositories/3996/datasets/ab13162732ae9ca1a9a27e2efd3da923ed6a4e7b/'
urlretrieve(url+file, file)
survey_data = pd.read_csv(file)

survey_data['Age Category'] = '21+'
idx = survey_data['Age'] < 21
survey_data.loc[idx, 'Age Category'] = 'Less than 21'

idx = survey_data['Mathematics'] >=4
survey_data['Interested in Math'] = False
survey_data.loc[idx, 'Interested in Math'] = True

lookup = {1: 'Never',
          2: 'Rarely',
          3: 'Sometimes', 
          4: 'Often',
          5: 'Always'}

for key, val in lookup.items():
    idx = survey_data["Parents' advice"] == key
    survey_data.loc[idx, 'Parents Advice'] = val

idx = survey_data['Siblings'] == 0
survey_data.loc[idx, 'Number of Siblings'] = '0'
idx = (survey_data['Siblings'] >= 1) & (survey_data['Siblings'] <= 2)
survey_data.loc[idx, 'Number of Siblings'] = '1 - 2'
idx = survey_data['Siblings'] >= 3
survey_data.loc[idx, 'Number of Siblings'] = '3+'

survey_data['Feels Lonely'] = False
idx = survey_data['Loneliness'] > 3
survey_data.loc[idx, 'Feels Lonely'] = True

survey_data['Interested in Pets'] = False
idx = survey_data['Pets'] > 3
survey_data.loc[idx, 'Interested in Pets'] = True

survey_data['Likes Techno'] = False
idx = survey_data['Techno'] > 3
survey_data.loc[idx, 'Likes Techno'] = True

survey_data.info()
```

```python
# Set the style to "whitegrid"
sns.set_style("whitegrid")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
sns.set_context('notebook')

file = 'mpg.csv'
url = 'https://assets.datacamp.com/production/repositories/3996/datasets/e0b285b89bdbfbbe8d81123e64727ff150d544e0/'
urlretrieve(url+file, file)
mpg = pd.read_csv(file)
mpg.info()
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

```python
mpg_mean = mpg[['model_year', 'origin', 'mpg']].groupby(['model_year', 'origin']).mean().reset_index()
mpg_mean = mpg_mean.rename({'mpg': 'mpg_mean'}, axis=1)
mpg_mean.info()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year",
      ylabel="Average MPG")

# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
# Set palette to "Blues"
sns.set_palette('Blues')

# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender",
                y="Age", data=survey_data, 
                kind="box", hue='Interested in Pets')

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not")

# Show plot
plt.show()
```

```python
# Set the figure style to "dark"
sns.set_style("dark")

# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Likes Techno", 
                data=survey_data, kind="bar",
                col='Gender')

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
      ylabel="% Who Like Techno")

# Show plot
plt.show()
```
