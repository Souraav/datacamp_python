```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd

# Load in the time series
candy = pd.read_csv('candy_production.csv', 
            index_col='date',
            parse_dates=True)

# Plot and show the time series on axis ax
fig, ax = plt.subplots()
candy.plot(ax=ax)
plt.show()
```

```python
# dicky-fuller tests whether a time series is stationary
earthquakes = pd.read_csv('earthquakes.csv',
                          index_col='date',
                          usecols=['date', 'earthquakes_per_year'],
                          parse_dates=True)
earthquakes['earthquakes_per_year'].plot()
plt.show()
```

```python
# Import augmented dicky-fuller test function
from statsmodels.tsa.stattools import adfuller

# Run test
result = adfuller(earthquakes['earthquakes_per_year'])

# Print test statistic
print(result[0])

# Print p-value
print(result[1])

# Print critical values
print(result[4])
```

```python
from urllib.request import urlretrieve

file = 'amazon_close.csv'
url = 'https://assets.datacamp.com/production/repositories/4567/datasets/4543d63de229cec637e58f90973b64417e5dc24c/'

urlretrieve(url+file, file)
amazon = pd.read_csv(file, index_col='date', parse_dates=True)

amazon.plot()
plt.show()
```

```python
amazon.info()
```

```python
# Calculate the first difference and drop the nans
amazon_diff = amazon.diff().dropna()

# Run test and print
result_diff = adfuller(amazon_diff['close'])
print(result_diff)
```

```python
# Calculate the first difference and drop the nans
amazon_diff = amazon.diff().dropna()

# Run test and print
result_diff = adfuller(amazon_diff['close'])
print(result_diff)

# Calculate log-return and drop nans
amazon_log = np.log(amazon/amazon.shift(1)).dropna()

# Run test and print
result_log = adfuller(amazon_log['close'])
print(result_log)
```
