Now pick the model.

Non-seasonal components:
* We took a 1 and 12 month diff. So for non-seasonal we have (p,1,q) - because we took a single diff.
* The non-seasonal acf and pacf's are close to not significant, and don't have any cut offs. And because it's not significant, we can't claim they both tail off - so we would say there are non-seasonal components. (0,1,0)

Seasonal Components:
* We took a 12 month diff, so that is our seasonal term (P, D, Q, 12)
* The seasonal acf cuts off after 1 term, and the pacf tails off. So we would say MA model order 1. (0,1,1,12) I'm not sure why d is 1 in this case - is it always so for a seasonal model?

So we could create it with these terms:
SARIMAX(order=(0,1,0), seasonal_order=(0,1,1,12))
